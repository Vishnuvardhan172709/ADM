```python
import pandas as pd
url='/content/netflix_titles ADM.csv'
data=pd.read_csv(url)
summary=data.describe()
print(summary)
```

```python
import matplotlib.pyplot as plt
plt.hist(data['title'].dropna(),bins=20,edgecolor='Black',color='skyblue')
plt.title('title Distribution')
plt.xlabel('title')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar chart for 'director' (Passenger class)
plt.figure(figsize=(6, 4))
data['director'].value_counts().plot(kind='bar', color='lightcoral')
plt.title('Passenger Class Distribution')
plt.xlabel('Class')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
# Box plot for 'release_year' to visualize outliers and distribution
plt.figure(figsize=(6,4))
plt.boxplot(data['release_year'].dropna(), patch_artist=True,
boxprops=dict(facecolor='lightgreen'))
plt.title('Box Plot of release_year')
plt.ylabel('release_year')
plt.show()
```

```python
import seaborn as sns
# Pairplot to visualize relationships between numerical features
sns.pairplot(data[['release_year']].dropna())
plt.show()
```

```python
# Correlation heatmap
correlation_matrix = data[['release_year']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', cbar=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Count plot for 'type' (how many survived vs how many did not)
sns.countplot(x='type', data=data, palette='Set2')
plt.title('type Count')
plt.xlabel('type')
plt.ylabel('Count')
plt.show()
```

```python
# Violin plot for 'type' across 'rating'
sns.violinplot(x='rating', y='type', data=data, palette='muted')
plt.title('type Distribution by Passenger Class')
plt.show()
```
